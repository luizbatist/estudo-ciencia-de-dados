# Criação de Vetores e Arrays em numpy

```python
import numpy as np
```

# ✔ Array

É o objeto principal do NumPy e pode ter 1, 2, 3 ou mais dimensões.

```python
minha_lista = [1, 2, 3] #vetor é uma array simples de 1 minha
np.array(minha_lista)
```

```python
minha_matriz = [1, 2, 3],[4, 5, 6] #array 2 dimensões é uma matriz
np.array(minha_matriz)
```

## E se eu quiser somar os elementos de uma matriz?

```python
matriz_numpy = np.array(minha_matriz)
np.sum(matriz_numpy)
```

## Criando um array de valores dentro de uma range 

```python
np.arange(0,10, 2) #parecido com for só que muito mais rápido
```

## Array com zeros

```python
np.zeros([3, 2]) #matriz com 3 dimensões e 2 colunas
```

```python
np.zeros([3, 3, 3]) #linha coluna e profundidade como se fosse um cubo 
```

## Criando matriz de identidade
utilizado muito em algebra linear e etc

```python
np.eye(4) #diagonal preenchida com o número 1
```

## criando uma sequencia de valores contínua igualmente espaçada

```python
np.linspace(0, 10, 5) #gera 5 valores igualmente espaçados
```

## Criação de valores aleatórios entre 0 e 1

```python
np.random.random(2)
```

```python
np.random.random(2) * 10 
#números aleatórios entre 0 e 10
```
