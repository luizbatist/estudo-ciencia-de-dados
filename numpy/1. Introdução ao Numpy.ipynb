# O que é NumPy?
O NumPy (Numerical Python) é uma biblioteca Python projetada para trabalhar com arrays multidimensionais e matrizes contendo números (valores numéricos). Essas estruturas de dados permitem representar eficientemente conjuntos de dados retangulares.

O NumPy possui várias vantagens, incluindo:

Velocidade aprimorada em operações matemáticas e manipulação de matrizes
Menor uso de memória para armazenar dados
Integração com outras bibliotecas populares como o Pandas e Scikit-Learn
Em resumo, o NumPy facilita muito o trabalho com grandes conjuntos de dados numéricos. Em vez de usar listas Python normais, podemos usar arrays NumPy que são mais rápidos e compactos.
# Por que usar NumPy?
Existem algumas razões importantes para usar o NumPy em projetos de análise de dados:

Velocidade: As operações com arrays NumPy podem ser muito mais rápidas do que com listas Python regulares. Isso porque os arrays NumPy são mais eficientes e permitem vetorização de operações.

Menos memória: Arrays NumPy ocupam menos memória para armazenar dados do que listas Python equivalentes. Isso permite trabalhar com conjuntos de dados maiores.

Conveniência: O NumPy possui uma ampla variedade de funções matemáticas e estatísticas embutidas para trabalhar com arrays numéricos.

Compatibilidade: O NumPy trabalha perfeitamente com muitas outras bibliotecas populares como Pandas, SciPy, Matplotlib, Scikit-Learn e mais. Usar arrays NumPy torna mais fácil a integração entre bibliotecas.

Portanto, se você vai trabalhar com conjuntos de dados numéricos médios ou grandes, o NumPy é uma ferramenta essencial.

# O que são Arrays?
Um array é o objeto central que o NumPy trabalha. Um array NumPy é uma grade retangular de valores (itens de dados), todos do mesmo tipo, acessíveis através de índices.

Em outras palavras, um array NumPy:

Tem um número fixo de itens dados
Todos os itens são do mesmo tipo primitivo (por exemplo, inteiros, floats ponto flutuante)
Os itens são acessados por índices como linhas e colunas
Os arrays NumPy têm vantagens sobre as listas Python:

Arrays são mais compactos e confiáveis
Operações matemáticas são muito mais rápidas com vetorização
Menos memória e mais velocidade
Dimensões fixas facilitam cálculos
Em resumo, arrays NumPy são a estrutura de dados principal para análise de dados rápida e de alto desempenho.

# Criando Arrays
Há algumas maneiras de criar novos arrays com NumPy:

## A partir de listas Python
Podemos converter listas Python para arrays NumPy usando np.array():

```python
num_list = [1, 2, 3] num_array = np.array(num_list)
```

## Arrays Multidimensionais
Podemos especificar o número de dimensões (linhas e colunas) de um array:

```python
mult_array = np.array([[1, 2, 3], [4, 5, 6]]) # 2x3 array
```

## Funções Construtoras
Funções como np.zeros(), np.ones() e np.full() permitem criar novos arrays rapidamente:

```python
empty_array = np.zeros((3, 4)) # 3x4 array de zeros
```

## Indexação e Fatiamento
Podemos acessar elementos individuais de arrays NumPy usando índices de linha e coluna:

```python
my_array = np.array([[1, 2, 3], [4, 5, 6]]) first_col = my_array[:,0] # Primeira coluna inteira second_row_first_col = my_array[1,0] # Segunda linha, primeira coluna
```
