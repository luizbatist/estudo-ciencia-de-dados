# O que é a biblioteca Pandas?
A Pandas é a biblioteca mais popular do Python para análise e manipulação de dados. O objetivo dela é tornar a análise de dados em código aberto mais poderosa, rápida e flexível do que usar planilhas como Excel ou Calc.

A Pandas trabalha com dados tabulares, como planilhas e tabelas. Também pode trabalhar com séries temporais, dados observacionais, estatísticos e matrizes arbitrárias por meio da integração com a biblioteca NumPy.

# Análise de dados com Pandas
Uma análise de dados com Pandas geralmente segue um fluxo:

Definir o objetivo central da análise (ex: entender o perfil de clientes de um banco)
Obter o conjunto de dados (CSV, Excel, XML, etc)
Ler e entender os metadados (o que cada coluna significa)
Entender a estrutura do conjunto de dados
Extrair informações específicas
Organizar e tratar os dados (remover duplicatas, valores faltantes, etc)
Análise exploratória para entender a distribuição dos dados
Extrair atributos e estatísticas dos dados
Responder às perguntas da análise


# Estruturas de dados do Pandas
O Pandas possui duas estruturas de dados principais:

Séries (Series): unidimensional, como um array rotulado. Possui apenas uma dimensão de índices e valores.

Dataframes: bidimensional, como uma tabela com linhas e colunas. Possui linhas e colunas rotuladas, podendo armazenar tipos de dados diferentes em cada coluna.
