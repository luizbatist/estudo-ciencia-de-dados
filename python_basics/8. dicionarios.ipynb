# O que São Dicionários?
*Imagine um dicionário de língua portuguesa. Cada palavra tem um significado associado, e é essa combinação de palavra (chave) e significado (valor) que torna o dicionário útil. Em Python, os dicionários funcionam de forma semelhante. Eles armazenam pares de chave-valor, onde cada chave é única e associa a um valor específico.*

```python
dicionario_vazio = {}
```

```python
dicionario_pessoa = {
    "idade" : 21,
    "nome" : "Luiz",
    "altura" : 1.67
}

dicionario_pet = {
    "nome": "Tica",
    "idade": 4,
    "cor": "preta"
}

print(f"Eu sou o {dicionario_pessoa['nome']} e tenho uma gatinha que se chama {dicionario_pet['nome']} tem a cor {dicionario_pet['cor']} e é muito mimada!")
```

```python
del dicionario_pet['cor']#deleta chaves que estejam presentes no dicionario 
print(dicionario_pet)
```

```python
tem_chave_cor = "cor" in dicionario_pet
print(tem_chave_cor)
```

```python
tem_chave_idade = 'idade' in dicionario_pet
print(tem_chave_idade)
```

```python
if 'altura' in dicionario_pessoa:
    print(dicionario_pessoa)
```

```python
dict_produto = {
    "nome": "cadeira",
    "preco": 237.50,                          #DICIONARIO CRIADO DO PRODUTO
    "estoque": 15
}

soma_lucro = 0  #SOMA DO LUCRO COMEÇANDO EM 0 POIS AINDA NÃO HOUVERAM VENDAS
while dict_produto['estoque'] > 0: #ENQUANTO FOR VERDADE A AFIRMAÇÃO BLOCO DE CÓDIGO ABAIXO SERÁ EXECUTADO
    print(f"Posso vender pois tenho {dict_produto['estoque']} em estoque!")   #EXIBE NA TELA QUE TEMOS EM ESTOQUE O PRODUTO 
    dict_produto['estoque'] -= 1 #CASO TENHA VENDA IRÁ DIMINUIR 1 PRODUTO NO ESTOQUE
    soma_lucro += dict_produto["preco"] #CASO TENHA A VENDA IRÁ SOMAR O VALOR DO PRODUTO NA VARIAVEL DE LUCRO, ASSIM DEIXANDO SALVO QUANTOS R$ GANHAMOS

print(dict_produto) #EXIBE O DICIONARIO DO PRODUTO NA TELA
print(f"Total de lucro R${soma_lucro:.2f}") #EXIBE O TOTAL DO LUCRO NA TELA, FILTRANDO EM 2 CASAS DECIMAIS
    
```
