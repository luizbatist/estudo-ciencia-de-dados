### Entendendo o While
*O while é um laço condicional que executa um bloco de código enquanto uma condição é verdadeira. Sua estrutura é semelhante ao if, mas com uma diferença crucial: o while continua executando o bloco de código enquanto a condição for verdadeira.*

*A sintaxe do while é a seguinte: while condição: # bloco de código*
# Exemplo Prático:

```python
contador = 5 
while contador >= 0: 
    print(contador) 
    contador -= 1
# Este exemplo imprime os números de 5 a 0, decrementando o contador a cada iteração.
```

# Diferenças entre For e While
*Embora ambos os laços executem código repetidamente, há diferenças significativas:*

*For: Ideal para iterar sobre sequências conhecidas, como listas ou strings.
While: Usado quando não se sabe previamente quantas vezes o loop será executado, dependendo de uma condição.*

# Controle do Fluxo com Break e Continue
*O while se torna ainda mais poderoso com as palavras-chave break e continue, que permitem controlar o fluxo de execução.*

# Break: Interrompendo o Loop
*break interrompe o loop imediatamente, saindo do bloco de código. Exemplo:*

```python
contador = 11 
while contador > 0: 
    if contador % 3 == 0: 
        break 
    print(contador) 
    contador -= 1
    # Este código imprime números de 11 a 1, interrompendo ao encontrar um múltiplo de 3.
```

# Continue: Pulando Iterações
*continue pula para a próxima iteração sem executar o restante do bloco. Exemplo:*

```python
contador = 11 
while contador > 0: 
    if contador % 3 == 0: 
        contador -= 1 
        continue 
    print(contador) 
    contador -= 1
# Aqui, números múltiplos de 3 são pulados, não sendo impressos.
```

### Aplicações Práticas do While
# 1. Soma de Números
*Um exemplo clássico é a soma de números até que uma condição seja atingida.*

```python
soma = 0 
numero = 1 
while True: 
    soma += numero 
    if soma > 100: 
        break 
    numero += 1 
print(f"Soma final: {soma}")
```

# 2. Imprimindo Números Pares
*Usando continue para imprimir apenas números pares*

```python
i = 1
while i <= 20:
    if i % 2 == 0:
        i += 1
        continue
    print(i)
    i += 1
```
