# O que são Funções?
*Antes de mergulharmos na sintaxe do Python, vamos entender o que são funções. No mundo da matemática, uma função é como uma caixa mágica. Você coloca um número dentro (o "x"), a caixa faz algum cálculo e devolve outro número (o "y"). Em programação, as funções funcionam de forma muito semelhante.*

*Uma função é um bloco de código que recebe entradas (argumentos), processa essas entradas e retorna uma saída. Ela é como uma rotina que pode ser executada várias vezes com diferentes entradas, tornando seu código mais flexível e reutilizável.*

# Por Que Usar Funções?
*Reutilização de Código: Em vez de reescrever o mesmo código várias vezes, você pode criar uma função e chamá-la sempre que necessário.
Organização: As funções ajudam a dividir o código em partes menores e mais fáceis de entender, tornando o programa mais organizado.
Manutenção: Se você precisar alterar uma parte do código, basta modificá-la dentro da função, e todas as chamadas a essa função serão automaticamente atualizadas*

```python
#EXEMPLOS
def soma (a, b): # def define uma função chamada soma, com dois parâmetros: a e b. 
    return a + b # return devolve o resultado da soma para quem chamou a função.
soma(1,5) # chamada da função com os argumentos 1 e 5 — executa e retorna 6.
```

```python
def celsius_para_fahrenheit(celsius):  # define uma função chamada celsius_para_fahrenheit com o parâmetro celsius
    return celsius * 9/5 + 32  # retorna o resultado da conversão de Celsius para Fahrenheit

celsius_para_fahrenheit(0)  # chamada da função com o argumento 0; retorna 32.0, equivalente a 0°C em Fahrenheit
```

```python
def saudacao(nome, cidade):  # define uma função chamada saudacao com dois parâmetros: nome e cidade
    print(f"Seja bem-vindo, {nome} de {cidade}")  # exibe a mensagem usando os valores recebidos; não retorna nada
saudacao("Luiz", "São Roque")  # chamada da função com os argumentos "Luiz" e "São Roque"
```

```python
def area_retangulo(largura, altura):
    area = largura * altura
    
    return area
l = 10
a = 6

area_retangulo(largura = l, altura = a)
```

```python
area_retangulo(l, a)
```

```python
area_retangulo(largura = 40, altura = 20)
```

```python
area_retangulo(altura= a, largura= l)
```

```python
def volume_cubo(largura, altura, comprimento):  # define a função volume_cubo com três parâmetros
    return largura * altura * comprimento  # retorna o resultado do cálculo do volume

# chamada posicional: os valores são passados na ordem dos parâmetros
volume_pos = volume_cubo(43, 5, 7)

# chamada nomeada: os valores são passados explicitamente com o nome de cada parâmetro
volume_nom = volume_cubo(largura=43, altura=5, comprimento=7)

# exibe os resultados das duas chamadas
print(volume_pos, volume_nom)
```

# Retornando valores de funções
## Verificando se número é par ou ímpar

```python
def numero_par(numero):
    if numero % 2 == 0:
        return("Par")
    else:
        return("Impar")

numero_par(12)
```
