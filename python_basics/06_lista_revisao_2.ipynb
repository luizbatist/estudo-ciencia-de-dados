```python
"""
Controle de notas escolares
Pe√ßa as notas de uma turma e calcule a m√©dia geral.
O programa deve continuar pedindo notas at√© que o usu√°rio digite -1.
No final, mostre a quantidade de alunos e a m√©dia da turma.

"""
soma = 0 #acumulador vai somar todas as notas
contador = 0 #vai contar quantas notas foram digitadas
while True: #la√ßo infinito enquanto for verdade, sai s√≥ com o break
    notas = float(input("Digite sua nota: ")) #vari√°vel guardando a nota digitada
    if notas == -1: #condicional afim de sair do loop
        break #parar a condi√ß√£o
    soma += notas #soma as notas no loop
    contador += 1 #conta a cada loop a nota digitada 

if contador > 0:
    media = soma / contador #calcula a m√©dia
    print(f"M√©dia das notas: {media:.2f}")
else:
    print("Nenhuma nota foi digitada...")
    
    
```

```python
"""
Caixa registradora simples
Simule o fechamento de um caixa:

O usu√°rio informa o valor de cada venda.

Quando digitar 0, o programa mostra o total vendido e o n√∫mero de vendas.

"""

soma = 0 #acumulador somando o valor das vendas
contador = 0 #conta quantas vendas foram feitas

while True: #loop infinito que s√≥ para com break
    venda = float(input("Qual foi o valor do produto?")) #vari√°vel guarda o valor da venda
    if venda == 0: #condi√ß√£o de parada (quando usu√°rio digita 0)
        break 
    soma += venda #acumula o valor dos produtos na vari√°vel soma
    contador += 1 #conta quantas vendas foram feitas e guarda na vari√°vel 
print(f"Total vendido: R$ {soma:.2f}") #exibe o total vendido e o numero de vendas
print(f"Total de vendas: {contador}")
print("\nCaixa encerrado at√© a pr√≥xima!")

if contador > 0:
    ticket_medio = soma / contador
    print(f"Ticket m√©dio √© de: R${ticket_medio:.2f}")
```

### üí∞ Bloco 2 ‚Äì C√°lculos e condi√ß√µes combinadas

```python
"""

C√°lculo de imposto de renda simplificado
Pe√ßa o sal√°rio e aplique as seguintes regras:

At√© R$2000 ‚Üí isento

De R$2001 a R$5000 ‚Üí 8%

Acima de R$5000 ‚Üí 15%
Mostre o valor do imposto e o sal√°rio l√≠quido.
"""

salario = "" #vari√°vel vazia antes de come√ßar o algoritmo
salario = float(input("Digite aqui o seu sal√°rio R$: ")) #solicita o salario ao usuario
if salario <= 2000: #condicional para calculo de i.r
    print("Isento de I.R") #mostra na tela o valor do i.r
    imposto_renda = 0 #vari√°vel guardando o valor do i.r
elif 2000 < salario <= 5000:#caso a primeira condi√ß√£o n√£o seja validada ser√° essa
    imposto_renda = salario * 0.08
    print(f"Seu I.R √© de R$: {imposto_renda:.2f}")
else: #caso as 2 condicionais n√£o forem validadas ser√° essa
    imposto_renda = salario * 0.15
    print(f"Seu I.R √© de R$: {imposto_renda:.2f}")
#calculo de salario l√≠quido
salario_liquido = salario - imposto_renda #calculo de salario - i.r

print(f"Seu sal√°rio l√≠quido √© de R$ {salario_liquido:.2f}") #exibe na tela o salario liquido
    
```
