### Desenvolvendo o Gerenciador de Tarefas
*Visão Geral do Projeto*
*Nosso gerenciador de tarefas incluirá as seguintes funcionalidades:*
*Adicionar tarefas: Permitirá que o usuário insira novas tarefas com uma descrição e um nível de prioridade.*
*Remover tarefas: Opção para excluir tarefas específicas.*
*Listar tarefas: Exibirá todas as tarefas atuais, junto com suas prioridades.*
*Salvar tarefas: Salvará as tarefas em um arquivo para que possam ser recuperadas posteriormente.*

# Etapa 1 do Case - Entrada e Saída de Dados com input() e print()

```python
tarefa = input("Digite uma tarefa: ")
print(f"Tarefa foi adicionada!")
print(f"Tarefas: {tarefa}")
```

# Etapa 2 do Case: Uso de Laços de Repetição e Condições

```python
while True:
    opcao = input("Digite uma tarefa (ou 'sair' para finalizar): ")
    if opcao.lower() == "sair":
        break
    tarefa = opcao
    print(f"Tarefa adicionada: {tarefa}")
print(f"Última tarefa adicionada: {tarefa}")
```

# Etapa 3 do Case: Estruturas Condicionais e Loops para Gerenciar Tarefas

```python
tarefas = []

while True:
    print("\n** Menu de opções **")
    print("1. Adicionar tarefa: ")
    print("2. Listar tarefas")
    print("3. Remover tarefa")
    print("4. Sair")

    opcao = input("Escolha uma opção: ")
    if opcao == "1":
        nome = input("Digite o nome da tarefa: ")
        prioridade = input("Digite a prioridade da tarefa: ")
        tarefa = {
            "nome": nome,
            "prioridade": prioridade
        }
        tarefas.append(tarefa)
        print(f"Tarefa adicionada: {tarefa}")
        
    elif opcao == "2":
        print("Lista de tarefas: ")
        for tarefa in tarefas:
            print(f"- {tarefa['nome']} (Prioridadede {tarefa['prioridade']})")
            
    elif opcao == "3":
        nome = input("Digite o nome da tarefa para ser removida: ")
        for tarefa in tarefas:
            if tarefa['nome'] == nome:
                tarefas.remove(tarefa)
                print(f"Tarefa {tarefa['nome']} removida com sucesso!")
                break
        else:
            print("Essa tarefa não existe!")
        
    elif opcao == "4":
        print("Saindo...")
        break
        
    else:
        print("\nOpção inválida!")
```

# Etapa 5 do Case: Organizando o Código com Funções

```python
tarefas = []
def adicionar_tarefa(nome,prioridade):
    tarefa = {
        "nome": nome,
        "prioridade": prioridade
        }
    tarefas.append(tarefa)
    print(f"Tarefa adicionada: {tarefa}")

def listar_tarefas():
    print("Lista de tarefas: ")
    for tarefa in tarefas:
        print(f"- {tarefa['nome']} (Prioridadede {tarefa['prioridade']})")
        
def remover_tarefas(nome):
    for tarefa in tarefas:
        if tarefa['nome'] == nome:
            tarefas.remove(tarefa)
            print(f"Tarefa {tarefa['nome']} removida com sucesso!")
            break
        else:
            print("Essa tarefa não existe!")
    
while True:
    print("\n** Menu de opções **")
    print("1. Adicionar tarefa: ")
    print("2. Listar tarefas")
    print("3. Remover tarefa")
    print("4. Sair")

    opcao = input("Escolha uma opção: ")
    if opcao == "1":
        nome = input("Digite o nome da tarefa: ")
        prioridade = input("Digite a prioridade da tarefa: ")
        adicionar_tarefa(nome,prioridade)
        
    elif opcao == "2":
        listar_tarefas()
            
    elif opcao == "3":
        nome = input("Digite o nome da tarefa para ser removida: ")
        remover_tarefas(nome)
        
    elif opcao == "4":
        print("Saindo...")
        break
        
    else:
        print("\nOpção inválida!")
```

# Etapa 6 do Case: Salvando e Carregando Tarefas em Arquivos JSON

```python
import json

tarefas = []

def salvar_tarefas ():
    with open("tarefas.json", "w") as arquivo:
        json.dump(tarefas, arquivo)
    print("Tarefas salvas com sucesso!")

def carregar_tarefas ():
    try:
        with open ("tarefas.json", "r") as arquivo:
            tarefas = json.load(arquivo)
        print("tarefas carregadas com sucesso!")
        return tarefas
    except:
        print("Nenhum arquivo de tarefas encontrado! Começando uma nova lista")
        return []

def adicionar_tarefa(nome,prioridade):
    tarefa = {
        "nome": nome,
        "prioridade": prioridade
        }
    tarefas.append(tarefa)
    print(f"Tarefa adicionada: {tarefa}")

def listar_tarefas():
    print("Lista de tarefas: ")
    for tarefa in tarefas:
        print(f"- {tarefa['nome']} (Prioridadede {tarefa['prioridade']})")
        
def remover_tarefas(nome):
    for tarefa in tarefas:
        if tarefa['nome'] == nome:
            tarefas.remove(tarefa)
            print(f"Tarefa {tarefa['nome']} removida com sucesso!")
            break
        else:
            print("Essa tarefa não existe!")

tarefas = carregar_tarefas()

while True:
    print("\n** Menu de opções **")
    print("1. Adicionar tarefa: ")
    print("2. Listar tarefas")
    print("3. Remover tarefa")
    print("4. Salvar tarefas")
    print("5. Sair")

    opcao = input("Escolha uma opção: ")
    if opcao == "1":
        nome = input("Digite o nome da tarefa: ")
        prioridade = input("Digite a prioridade da tarefa: ")
        adicionar_tarefa(nome,prioridade)
        
    elif opcao == "2":
        listar_tarefas()
            
    elif opcao == "3":
        nome = input("Digite o nome da tarefa para ser removida: ")
        remover_tarefas(nome)

    elif opcao == "4":
        salvar_tarefas()
        
    elif opcao == "5":
        print("Saindo...")
        break
        
    else:
        print("\nOpção inválida!")
```
