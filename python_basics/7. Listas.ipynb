### O que São Listas em Python?
*lista é uma coleção ordenada de elementos. Esses elementos podem ser de qualquer tipo: números, strings, booleanos, ou até mesmo outras listas. O que torna as listas tão úteis é que elas permitem armazenar múltiplos valores em uma única variável, facilitando a manipulação e o acesso a esses dados.*

*Uma das características mais distintas de uma lista em Python é o uso de colchetes [] para definir seus elementos. Por exemplo, se quisermos criar uma lista com os números 3, 6 e 9, podemos fazer isso da seguinte maneira:*

```python
minha_lista = [3, 6, 9]
# Quando imprimirmos minha_lista, obteremos como saída:
print(minha_lista)
lista_variada = [0, "abacaxi", 3, True]

#somando elementos de listas diferentes
print(minha_lista[0] + lista_variada[2])
```

```python
#utilizando algortimo de condicional para verificação de número par ou ímpar dentro de uma lista
nova_lista = [1, 6, 7, 29, 3]
if nova_lista[1] % 2 == 0:
    print(f"O elemento '1' da lista é o número {nova_lista[1]} e é par")
else:
    print(f"O elemento '1' da lista é o número {nova_lista[1]} e é impar")
```

## Modificando Listas
# Adicionando Elementos a uma Lista
*Uma das operações mais comuns ao trabalhar com listas é adicionar elementos. Python oferece métodos convenientes para isso, como append e insert.*

# Utilizando o Método append
*O método append é usado para adicionar um elemento ao final de uma lista. Ele é simples e eficiente, tornando-se uma ferramenta indispensável para construir listas dinamicamente*
*Exemplo:*

```python
nums = [1, 2, 3, 4] 
nums.append(5) 
print(nums)
```

# Utilizando o Método insert
*Quando precisamos adicionar um elemento em uma posição específica, utilizamos o método insert. Ele recebe dois parâmetros: o índice onde o elemento será inserido e o próprio elemento.*

*Exemplo:*

```python
nums = [1, 2, 3, 4] 
nums.insert(0, 3) 
print(nums)
```

## Removendo Elementos de uma Lista
*Assim como adicionar, remover elementos é uma operação frequente. Python fornece métodos como remove, pop e clear para isso.*

# Utilizando o Método remove
*O método remove elimina a primeira ocorrência de um valor especificado. Se o valor não existir na lista, um erro será gerado.*
*Exemplo:*

```python
nums = [1, 2, 3, 4] 
nums.remove(1) 
print(nums)
```

# Utilizando o Método pop
*O método pop remove o elemento da posição especificada. Se nenhum índice for fornecido, ele remove o último elemento. Além disso, pop retorna o valor removido, o que pode ser útil em certas situações.*

*Exemplo:*

```python
nums = [1, 2, 3, 4] 
removido = nums.pop() 
print(nums)
```

# Utilizando o Método clear
*Para limpar todos os elementos de uma lista, utilizamos o método clear.*
*Exemplo:*

```python
nums = [1, 2, 3, 4] 
nums.clear() 
print(nums)
```

```python
#EXERCÍCIO PROPOSTO EM AULA
numeros = [10, 20, 30, 40, 50]
numeros.append(60) #INSERIR UM VALOR NO FINAL
numeros.insert(1,15) #INSERIR UM VALOR NA 2º POSIÇÃO
numeros.remove(30) #ELIMINANDO VALOR DE NÚMERO '30'
print(numeros)
```

```python
itens = ["Mesa", "Cadeira", "Poltrona"]
itens[0] = "Guarda Roupa"
removido = itens.pop()

print(f"Exibindo a lista com o elemento [0] modificado e o elemento [2] removido {itens}")
print(f"Removido o último elento da lista: {removido}")
```

### Introdução ao Slicing
*O slicing é uma técnica poderosa em Python que permite "fatiar" uma lista, selecionando apenas uma parte dela. Isso é útil quando você precisa trabalhar com um subconjunto de dados sem alterar a lista original.*

*Como Funciona o Slicing
O slicing é realizado usando colchetes [] com dois pontos : para indicar o início e o fim do subconjunto desejado. A sintaxe básica é:*

*sublista = lista[início:fim]*

exemplo prático:

```python
lista = [0, 1, 2, 3, 4, 5, 6]
sub_lista = lista[2:5] #EXIBIU O ELEMENTO 2 ATÉ O ELEMENTO 5
```

```python
len(lista) #CONTA QUANTOS ELEMENTOS TEM A LISTA
```

```python
lista.count(3) # QUANTOS ELEMENTOS TEM O NÚMERO 3
```

```python
outra_lista = [1, 2, 3, 3, 3, 2, 1]
outra_lista.count(3)
```

```python
lista.reverse() #INVERTE A LISTA
print(lista) 
```

### Percorrendo Listas com Loops
*As listas são fundamentais para loops em Python. Você pode usar loops for para percorrer os elementos de uma lista e executar operações neles.*

*Como Percorrer uma Lista*
*A sintaxe básica é:*

*for elemento in lista: # Faça algo com o elemento*

*Exemplo Prático:*

```python
for i in lista: #PERCORRE TODOS OS ELEMENTOS DA LISTA
    print(i)
```

```python
nums = [1, 20, 3, 4, 29, 18, 3]
soma = 0
for n in nums:       #MINHA SOMA PERCORREU ELEMENTOS DE 'NUMS' E SOMOU TODOS ELES
    soma += n
    print(soma)
```

```python
lista_frutas = ["banana", "manga", "ameixa", "melão", "banana", "abacaxi", "banana"] 
while "banana" in lista_frutas: #ENQUANTO ESSA AFIRMAÇÃO FOR VERDADE IRÁ RODAR O BLOCO DE CÓDIGO ABAIXO. Enquanto a palavra 'banana' estiver na lista
    lista_frutas.remove("banana") #executa enquanto a afirmação a cima for TRUE. depois de remover o python volta no inicio do loop e avalia novamente a condição
print(lista_frutas)
```
