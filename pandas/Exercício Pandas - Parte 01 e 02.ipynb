## Qual é o BasePay médio?

```python
df["BasePay"]
```

"Note Provided" está sendo tratado como String pois não tem como transformar em número, porque é um texto, isso está atrapalhando a base de dados.

```python
df = df[~(df["BasePay"] == "Not Provided")] #Pegue todas as linhas onde BasePay NÃO é "Not Provided". Exclua o resto.
```

```python
df["BasePay"] = df["BasePay"].astype(float) # Pegue a coluna BasePay, converta para float e SALVE o resultado de volta nessa coluna.
```

```python
df.info()
```

## Qual é o maior OvertimePay do dataset?

```python
df["OvertimePay"] = pd.to_numeric(df["OvertimePay"], errors="coerce")
```

```python
df["OvertimePay"].astype(float).max()
```

```python
valor_maximo = df["OvertimePay"].astype(float).max()
```

```python
df[df["OvertimePay"]==valor_maximo]
```

## Quantas pessoas têm a palavra chefe em seu cargo? (complicado)

```python
def chefe(x):
  if "chief" in x.lower():
    return True
  else:
    return False

df[df["JobTitle"].apply(chefe)]
```
