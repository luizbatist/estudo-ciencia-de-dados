```python
import numpy as np
import pandas as pd
```

```python
dados = {
    "Nome": ["Ana", "Bruno", "Carla", "Daniel", "Eduardo", "Fernanda"],
    "Departamento": ["TI", "TI", "RH", "RH", "Vendas", "Vendas"],
    "Salario": [6000, 6000, 4000, 4200, 3000, 3500],
    "Estado": ["SP", "RJ", "SP", "RJ", "RJ", "SP"]
}
```

```python
df = pd.DataFrame(dados)
```

```python
df
```

```python
#CALCULAR A MEDIA POR GROUPY BY DEPARTAMENTO
df.groupby("Departamento")["Salario"].mean()
```

```python
#CALCULAR O VALOR MAX POR GROUPY BY DEPARTAMENTO
df.groupby("Departamento")["Salario"].max()
```

```python
#CALCULAR O VALOR MIN POR GROUPY BY DEPARTAMENTO
df.groupby("Departamento")["Salario"].min()
```

```python
#MEDIA SALARIAL POR ESTADO
df.groupby("Estado")["Salario"].mean()
```

```python
#FAZENDO GROUP BY DE VARIOS CALCULOS JUNTOS, MÃ‰DIA, MAXIMO, MINIMO, COUNT, VALORES UNICOS
df.groupby("Departamento").agg({"Salario": ["mean", "median", "max", "min", "count", "nunique"]})
```

```python
df.groupby("Estado").agg({"Salario": ["mean", "max", "median", "min"]})
```
